## 获取TF网络

```r
import loompy
import pandas as pd
```

```r
def extract_tf_targets_from_loom(loom_file):
    with loompy.connect(loom_file, validate=False, mode='r') as ds:
        regulon_df = pd.DataFrame(ds.ra.Regulons)
        regulon_df.index = ds.ra.Gene

    return regulon_df
```

```r
df = read.csv("/data/work/fig3_TF/SCENIC_240806/tf_auc0.1.csv.gz", row.names=1)
colnames(df) = sub("\\.\\.\\.", "", colnames(df))
```

```r
stage_colors <- c("Fascia" = "#db2400", "Inflammatory" = "#63FFAC", "Oxidative Stress" = "#FFFF52", "Myofibroblast" = "#1CE6FF")
```

```r
plotTFgraph = function(TF, group){
    if (group == "Adult"){
        df = df
        mat = adult_mat
    } else {
        df = df
        mat = ep_mat
    }
    targets = rownames(df)[which(df[,TF]>0)]
    cat("Number of targets of ", TF, " (", group,"):", length(targets), "\n")


    nodes <- data.frame(name = unique(c(TF, targets)))
    nodes$correlation = sapply(nodes$name , function(x) cor(mat[x,], mat[TF, ])) 
    nodes$correlation[nodes$correlation<0] = 0
    nodes$pseudotime = apply(mat[nodes$name,],1,function(x){which.max(smooth.spline(x,df=3)$y)})   
    nodes$stages = c("Fascia", "Inflammatory", "Oxidative Stress", "Myofibroblast")[(nodes$pseudotime-1) %/% 30 + 1]
    
    edges = nodes[nodes$name %in% targets, ]
    edges$from = TF
    edges$to = edges$name
    edges$group = group                                                      
                               
                          
    Tgraph <- tbl_graph(nodes = nodes, edges = edges, directed = F)
    p = ggraph(Tgraph, layout = "stress") +   
      geom_edge_link(aes(alpha = correlation), color = "white") +  
      geom_node_point(aes(size = correlation, fill = stages), color="white", shape=21) +
      geom_node_text(aes(label = name, fontface = ifelse(name == TF, "bold", "plain")), 
                     size = 4, segment.colour = "white",
                     colour = "white", 
                     min.segment.length = 2,
                     repel = TRUE) +  
      theme_void() +  
      theme(
        plot.background = element_rect(fill = "black", color = NA),
        panel.background = element_rect(fill = "black", color = NA),
        legend.background = element_rect(fill = "black", color = NA),
        legend.key = element_rect(fill = "black", color = NA),
        text = element_text(color = "white")
      ) +  
      scale_edge_width(range = c(0.3, 2)) +
      scale_size_continuous(range = c(0.5, 5)) +
      scale_fill_manual(values = stage_colors) +
      # scale_color_viridis_c(option = "viridis", limits = c(0, 120)) +
      ggtitle(paste0(TF, " (", group, ")")) +
      geom_blank()
    list(p = p, edges = edges)
}
```

```r
dir.create("/data/work/fig3_TF/SCENIC_240806/graph_auc0.1")
result_df = data.frame()
for (tf in colnames(df)){
    result = plotTFgraph(tf, "Adult")
    ggsave(paste0("/data/work/fig3_TF/SCENIC_240806/graph_auc0.1/", "Adult_", tf, ".png"), result$p,width=8,height=7)
    result_df = rbind(result_df, result$edges)
}
for (tf in colnames(df)){
    result = plotTFgraph(tf, "EP")
    ggsave(paste0("/data/work/fig3_TF/SCENIC_240806/graph_auc0.1/", "EP_", tf, ".png"), result$p,width=8,height=7)
    result_df = rbind(result_df, result$edges)
}
write.csv(result_df,"/data/work/fig3_TF/SCENIC_240806/graph_auc0.1/TF_graph.csv")
```
